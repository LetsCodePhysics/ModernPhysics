## Theory

We saw in Assignment 6 that a particle's **wave function** $\psi$ is governed by the Schrodinger equation:

$$
-\frac{\hbar^2}{2m} \psi'' + V \psi = E \psi,
$$

with the important caveat that the energy $E$ had to be a special **eigenenergy** $E_n$ and that the wave function $\psi$ had to be a special **eigenfunction** $\psi_n$. Each $\psi_n$ is specifically paired with an $E_n$.

This was a true statement, but was a bit of a simplification. In reality, a **general wave function** $\Psi$ can be a **superposition** (a combination) of all the possible eigenfunctions, and **evolves** in time. Mathematically, we write this like so:

$$
\Psi(x,t) = c_1 \psi_1(x) e^{-i E_1 t / \hbar} + c_2 \psi_2(x) e^{-i E_2 t / \hbar} + c_3 \psi_3(x) e^{-i E_3 t / \hbar} + \ldots
$$

A few important items to note about this construction:
* The $c_n$'s are constants (called the **coefficients**) that determine how much weighting each eigenfunction $\psi_n$ receives. These values are constant, selected at the beginning of a problem. If a particular $c$-value equals zero, then that eigenfunction doesn't appear in the wave function (ever).
* The dependence on $x$ and $t$ are **separated**. Only the $\psi_n$'s depend on $x$ and only the exponential factors depend on $t$. Position and time never directly combine with each other. This separation is why we were interested in the eigenfunctions to begin with.
* The eigenenergies $E_n$ show up in the exponents. They determine how each piece of the wave function time-evolves.
* The exponential factors are **complex**; they include the imaginary number $i = \sqrt{-1}$. This may seem bizarre at first; that's why we need **Euler's formula**:
$$
e^{i\theta} = \cos(\theta) + i \sin(\theta).
$$
* This means that the general wave function $\Psi$ has a **real part** (based on the cosines) and an **imaginary part** (based on the cosines) that both oscillate with time. We'll look at each part individually.
