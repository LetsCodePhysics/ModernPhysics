We can simulate the xray diffraction pattern produced by a known latice structure to help us match experiment with theory. 

First, we specify a lattice structure by specifying three parameters: **edge lengths** $a$, $b$, and $c$, and **angles** between the edges $\alpha$, $\beta$, and $\gamma$ as depicted here:


For example, graphite (pictured a few cells above) has $a = b = 2.46 \times 10^{-10}$ meters, $c = 6.71 \times 10^{-10}$ meters, $\alpha = \beta = 90^\circ$, and $gamma = 120^\circ$.

Second, we define **primitive vectors** $\vec{a}$, $\vec{b}$, and $\vec{c}$ along each edge of the crystal lattice. The planes that we detect in xray diffraction are based on these vectors.

Third, we define **reciprocal vectors** as
$$ \begin{eqnarray}
\vec{a}^* & = & 2\pi \frac{\vec{b}\times\vec{c}}{\vec{a}\cdot\vec{b}\times\vec{c}} \\
\vec{b}^* & = & 2\pi \frac{\vec{c}\times\vec{a}}{\vec{a}\cdot\vec{b}\times\vec{c}} \\
\vec{c}^* & = & 2\pi \frac{\vec{a}\times\vec{b}}{\vec{a}\cdot\vec{b}\times\vec{c}}. 
\end{eqnarray} $$

These reciprocal vectors are defined such that $\vec{a}^* \cdot \vec{b} = \vec{a}^* \cdot \vec{c} = 0$, etc., even though $\vec{a}$, $\vec{b}$, and $\vec{c}$ don't have to be orthogonal.

Fourth, each plane in the crystal lattice that produces xray diffraction has a **normal vector** given by
$$
\hat{n} = h \vec{a}^* + k \vec{b}^* + l \vec{c}^*,
$$
where the integers $h$, $k$, and $l$ are the plane's **Miller indices**. We denote a plane in a crystal lattice by $(hkl)$, its triplet of Miller indices. 

Finally, because the unit cell repeats throughout the solid, each $(hkl)$ plane repeats throughout the solid, separated by a distance given by (deep breath)

$$
d = \sqrt{ \frac{1 - \cos^2(\alpha) - \cos^2(\beta) - \cos^2(\gamma) + 2 \cos(\alpha) \cos(\beta) \cos(\gamma)} {\frac{h^2}{a^2} \sin^2(\alpha) + \frac{k^2}{b^2}\sin(\beta)^2 + \frac{l^2}{c^2} \sin^2(\gamma) + \frac{2kl}{bc} (\cos(\beta) \cos(\gamma)-\cos(\alpha)) + \frac{2hl}{ac} (\cos(\alpha) \cos(\gamma)-\cos(\beta)) + \frac{2kh}{ba} (\cos(\beta) \cos(\alpha)-\cos(\gamma))} }
$$

I don't expect you to memorize this or understand where it comes from. But please check to make sure I've typed it up correctly in the code cells below! Here's the important part: **This is the $d$ that appears in the xray diffraction condition**
$$
2d\sin(\theta) = n \lambda.
$$

```python
from pylab import *
%matplotlib inline

# Set lattice parameters.
# a,b,c = edge lengths
# alpha, beta, gamma = angle between edges b&c, c&a, a&b
# If alpha, beta, gamma = 90 degrees, then a, b, & c
# point along x, y, & z axes.
alpha = 90
beta = 90
gamma = 90
a = 2.5
b = 2.5
c = 2.5

# Convert degrees to radians.
alpha *= pi/180
beta *= pi/180
gamma *= pi/180

# Calculate primitive vectors along each edge.
# a always points along x-axis.
aVec = array([a,0,0])
# b always points in xy-plane.
bVec = array([b*cos(gamma),b*sin(gamma),0])
# c goes elsewhere.
cx = c*cos(beta)
cy = c*(cos(alpha)-cos(beta)*cos(gamma))/sin(gamma)
cz = sqrt(c**2-cx**2-cy**2)
cVec = array([cx,cy,cz])

# Calculate reciprocal vectors.
aStar = 2*pi*cross(bVec,cVec)/dot(aVec,cross(bVec,cVec))
bStar = 2*pi*cross(cVec,aVec)/dot(aVec,cross(bVec,cVec))
cStar = 2*pi*cross(aVec,bVec)/dot(aVec,cross(bVec,cVec))
```

```python
xResults = [] # x-coordinate of each bright spot
yResults = [] # y-coordinate of each bright spot
zscreen = 100 # Distance to screen, along z-axis.
lam = 0.4*min(a,b,c) # Wavelength of xrays. Should be <~ interplanar distance.

inBeam = array([0,0,1]) # Direction of incoming beam.
xhat = array([1,0,0]) # Define xhat.
yhat = array([0,1,0]) # Define yhat.

# Loop over the (hkl) planes. Can expand range to attempt to catch more points.
for h in range(-10,10):
    for k in range(-10,10):
        for l in range(-10,10):
            if (h!=0 or k!=0 or l!=0): # Avoid dividing by zero.
                # Calculate unit vector normal to (hkl) plane.
                normalVector = h*aStar + k*bStar + l*cStar
                # Calculate interplanar distance d.
                num = 1 - cos(alpha)**2 - cos(beta)**2 - cos(gamma)**2 + 2*cos(alpha)*cos(beta)*cos(gamma)
                denom = h**2/a**2 * sin(alpha)**2 + k**2/b**2 * sin(beta)**2 + l**2/c**2 * sin(gamma)**2 + 2*k*l/(b*c)*(cos(beta)*cos(gamma)-cos(alpha)) + 2*h*l/(a*c)*(cos(alpha)*cos(gamma)-cos(beta)) + 2*k*h/(b*a)*(cos(beta)*cos(alpha)-cos(gamma))
                d = sqrt(num/denom)
                # Evaluate maximum n to avoid taking arcsin of a number > 1.
                nmax = int(2*d/lam)
                # Loop over n values that give constructive interference.
                for n in range(0,nmax):
                    # Calcualte location of bright spot. The 2 outside ensures we measure from the beam's axis.
                    theta = 2*arcsin(n*lam/(2*d))
                    # Calculate where the outgoing beam strikes the screen and record coordinates.
                    outBeam = (inBeam + normalVector*sin(theta))*zscreen
                    xResults.append(dot(outBeam,xhat))
                    yResults.append(dot(outBeam,yhat))
# Plot results. Make graph square to make angles meaningful.
scatter(xResults,yResults)
axis("square")
show()
```
